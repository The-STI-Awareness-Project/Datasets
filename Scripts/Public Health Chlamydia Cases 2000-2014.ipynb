```python
import re

# Read the CSV file into a DataFrame
df = pd.read_csv('../Public Health Chlamydia Cases 2000-2014.csv')

# Create a list to store dictionaries
data_list = []

# Regular expression pattern to extract the year
year_pattern = r'\b\d{4}\b'

# Iterate through rows and create dictionaries
for index, row in df.iterrows():
    for column in df.columns:
        if 'Cases' in column:
            # Extract the year using a regular expression
            year_match = re.search(year_pattern, column)
            if year_match:
                year = year_match.group()
                data_dict = {
                    'Community Area': row['Community Area'],
                    'Community Area Name': row['Community Area Name'],
                    'Year': year,
                    'Female 15-44': row[f'Cases {year} Female 15-44'],
                    'Incidence Rate Lower CI': row[f'Incidence Rate {year} Lower CI'],
                    'Incidence Rate Upper CI': row[f'Incidence Rate {year} Upper CI']
                }
                data_list.append(data_dict)

# Now, data_list contains dictionaries for each row and year
```

```python
data = pd.DataFrame(data_list)
data
```

```python
data.to_csv('../Public Health Chlamydia Cases 2000-2014.csv')
```
